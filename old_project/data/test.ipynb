```python
%matplotlib inline
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import community as community_louvain
import random
import time
import seaborn as sns
random.seed(1234)
from bokeh.io import output_notebook, show, save
output_notebook()

df_nodes = pd.read_csv('nodes.csv', sep=',')
df_edges = pd.read_csv('edges.csv', sep=',')

graph = nx.Graph()
graph.add_nodes_from(df_nodes['#BIOGRID ID'])
graph.add_edges_from(df_edges[['BioGRID Gene ID', 'Related BioGRID Gene ID']].values.tolist())
dict_nodes = df_nodes.set_index('#BIOGRID ID').to_dict()
for key in dict_nodes:
    nx.set_node_attributes(graph, dict_nodes[key], name=key)
dict_edges = df_edges.set_index(['BioGRID Gene ID', 'Related BioGRID Gene ID']).to_dict()
for key in dict_edges:
    nx.set_edge_attributes(graph, dict_edges[key], name=key)
graph.remove_node('-')
```

```python
pos = nx.spring_layout(graph, seed=4321, k=3)

fig = plt.figure(figsize=(20, 10))

nx.draw_networkx_nodes(graph, pos)
nx.draw_networkx_edges(graph, pos, alpha=0.1)
nx.draw_networkx_labels(graph, pos)
plt.show()
```
